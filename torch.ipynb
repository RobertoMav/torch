```python
X: torch.tensor = torch.randn([500, 1], device=device)
error: torch.tensor = torch.randn([500, 1], device=device) * 3
y: torch.tensor = X * 7 + 2 + error
```

```python
print(X.shape)
print(y.shape)
print(error.shape)
```

```python
linear_reg: torch.nn.Linear = torch.nn.Linear(in_features=1, out_features=1, device=device)
```

```python
BATCH_SIZE: int = 32
trainig_data: int = X.shape[0]
number_of_iter: int = trainig_data // BATCH_SIZE
```

```python
epochs: int = 100
learning_rate: float = 1e-3
for epoch in range(epochs):
    for i in range(500):
        preds = linear_reg(X[i])
        loss = torch.mean((preds - y[i])**2)

        # Compute gradients
        loss.backward()

        # Manually update parameters
        with torch.no_grad():
            for param in linear_reg.parameters():
                param -= learning_rate * param.grad
                param.grad.zero_()  # Clear gradients for next iteration
```
